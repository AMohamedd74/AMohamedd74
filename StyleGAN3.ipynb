```python
from PIL import Image
import os

source_dir = '/content/dataset/brain_tumor'
output_dir = '/content/dataset/brain_tumor_resized'

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Iterate over the images in the source directory
for filename in os.listdir(source_dir):
    if filename.endswith(".jpg") or filename.endswith(".png") or filename.endswith(".JPG"):
        # Open the image
        image_path = os.path.join(source_dir, filename)
        image = Image.open(image_path)

        # Resize the image to 512x512
        resized_image = image.resize((512, 512))
        image_rgb = resized_image.convert("RGB")

        # Save the resized image to the output directory
        output_path = os.path.join(output_dir, filename)
        image_rgb.save(output_path)

        print(f"Resized and saved {filename} to {output_path}")
```
